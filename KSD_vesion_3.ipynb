### Изменение графиков стабильности температуры

```python
def change_temperature_stab_graph(value):
    square = 0
    square_norm = 0
    trace_list = []
    yx = []
    max_h = np.max(initial_data[list(sorted_dict.keys())[value[0]]]["Глубина"])
    ind_max_h = list(initial_data[list(sorted_dict.keys())[value[0]]]["Глубина"]).index(max_h)
    xp = np.linspace(5 /max_h,np.max(initial_data[list(sorted_dict.keys())[value[0]]]["Глубина"])/ max_h,41)
    
    
    
    
    
    
    interp_base = np.interp(xp,[initial_data[list(sorted_dict.keys())[value[0]]]["Глубина"][i] / initial_data[list(sorted_dict.keys())[value[0]]]["Глубина"][ind_max_h]
                                                     for i in range(len(initial_data[list(sorted_dict.keys())[value[0]]])) 
                                                     if (initial_data[list(sorted_dict.keys())[value[0]]]["Параметр_положения"][i] == 1 
                                                     and initial_data[list(sorted_dict.keys())[value[0]]]["Глубина"][i] > 2)][::-1],
                           [initial_data[list(sorted_dict.keys())[value[0]]]["Температура"][i] / initial_data[list(sorted_dict.keys())[value[0]]]["Температура"][ind_max_h]
                                                     for i in range(len(initial_data[list(sorted_dict.keys())[value[0]]])) 
                                                     if (initial_data[list(sorted_dict.keys())[value[0]]]["Параметр_положения"][i] == 1 
                                                     and initial_data[list(sorted_dict.keys())[value[0]]]["Глубина"][i] > 2)][::-1])
    for key in range(value[0],value[1]+1):
        interp = np.interp(xp,[initial_data[list(sorted_dict.keys())[key]]["Глубина"][i] / initial_data[list(sorted_dict.keys())[value[0]]]["Глубина"][ind_max_h]
                                                     for i in range(len(initial_data[list(sorted_dict.keys())[key]])) 
                                                     if (initial_data[list(sorted_dict.keys())[key]]["Параметр_положения"][i] == 1 
                                                     and initial_data[list(sorted_dict.keys())[key]]["Глубина"][i] > 2)][::-1],
                           [initial_data[list(sorted_dict.keys())[key]]["Температура"][i] / initial_data[list(sorted_dict.keys())[value[0]]]["Температура"][ind_max_h]
                                                     for i in range(len(initial_data[list(sorted_dict.keys())[key]])) 
                                                     if (initial_data[list(sorted_dict.keys())[key]]["Параметр_положения"][i] == 1 
                                                     and initial_data[list(sorted_dict.keys())[key]]["Глубина"][i] > 2)][::-1])
        square += np.sum(((interp_base - interp)) ** 2)/41
        square_norm += np.sum(((interp_base - interp)/interp_base) ** 2)
        yx.append(square)
    if (value[1] != value[0]):
        #square = square / (41) ## 41 - кол-во точек
        square_norm = 1 - square_norm
    else:
        square = 0
        square_norm = 0
    trace_list.append(go.Scatter(visible=True, x = [i for i in range(len(yx))], y=[1-i for i in yx]))
    traces = [trace_list]
    data = [val for sublist in traces for val in sublist]
    figure = {'data': data,
              'layout': go.Layout(
                  paper_bgcolor='white',
                  plot_bgcolor='white',
                  margin={'b': 50, 'r': 0, 'l': 50, 't': 10 },
                  height = 350,
                  autosize=True,
                  #title={'text': 'Зависимость давления', 'font': {'color': 'Black'}, 'x': 0.55},
                  xaxis={'title':'Data','tickcolor': 'black',  "linecolor":'black' },
                  yaxis={'title':'Pressure','tickcolor': 'black', "linecolor":'black' },
              ),

              }
    return ['Нормированное cреднее квадратичное отклонение от первой кривой спуска "{:.3f}"'.format(1-square)], figure
```

### Изменение графиков по срезу температуры

```python
def change_temperature_const_h_graph(value_vert, value_hor):
    trace_list = []
    requiredNumber = 0;
    yx=[]
    for key in range(value_hor[0],value_hor[1]+1):
        i = 0
        while i <= len(initial_data[list(sorted_dict.keys())[key]]):
            if value_vert <= initial_data[list(sorted_dict.keys())[key]]['Глубина'][i]:
                requiredNumber = i
                break
            i +=1
        yx.append(initial_data[list(sorted_dict.keys())[key]]["Температура"][requiredNumber])
    trace_list.append(go.Scatter(visible=True, x = [i for i in range(len(yx))], y=[i for i in yx],line=dict(color='red')))
    traces = [trace_list]
    
    data = [val for sublist in traces for val in sublist]
    figure = {'data': data,
              'layout': go.Layout(
                  paper_bgcolor='white',
                  plot_bgcolor='white',
                  margin={'b': 50, 'r': 0, 'l': 50, 't': 10 },
                  height = 350,
                  autosize=True,
                  #title={'text': 'Зависимость давления', 'font': {'color': 'Black'}, 'x': 0.55},
                  xaxis={'title':'Data','tickcolor': 'black',  "linecolor":'black' },
                  yaxis={'title':'Temperature','tickcolor': 'black', "linecolor":'black' },
              ),

              }

    return figure
```
